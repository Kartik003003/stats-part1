Measure of Dispersion
1. variance
2. standard deviation

```python
st.variance(age)    # uses sample variance
```

```python
np.var (age,axis=0)   # were/n is done
```

```python
def variance(data):
  n=len(age)
  mean=sum(data)/n
  deviation=[(x-mean)**2 for x in data]
  variance=sum(deviation)/n
  return variance
```

```python
variance(age)
```

```python
def varian(data,dof=0):
  n=len(age)
  mean=sum(data)/n
  deviation=[(x-mean)**2 for x in data]
  variance=sum(deviation)/(n-dof)
  return variance
```

```python
varian(age)
```

Check Wether Distribution is normal distribution

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample,kde=True)
  plt.subplot(1,2,2)
  stat.probplot(sample,dist='norm',plot=pylab)
  plt.show()
```

```python
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])
```

```python
def plot_data_log(sample):
  plt.figure(figsize=(10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample,kde=True)
  plt.subplot(1,2,2)
  stat.probplot(np.log(sample),dist='norm',plot=pylab)
  plt.show()
```

```python
plot_data_log(df['sepal_width'])
```
